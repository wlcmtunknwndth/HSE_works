```python
# Ð—Ð°Ð´Ð°Ð½Ð¸Ðµ 3
def format(s):
    if type(s) is not str:
        s = str(s)
    s = s.strip().replace(',', '.').strip('$').replace(' ', '')
    # s = s.replace(',', '.')
    # s = s.strip()
    # print(s)
    return float(s)
print(format(" 29,00 $ "))
# df['balance']
```

```python
# df = df['balance'].apply(format)
df['balance'] = df['balance'].apply(format)
display(df)
round(df['balance'].mean(), 3)
```
